```python

def rle2mask(mask_rle, shape):
    # эта функция взята с конкурса и я её оставил как есть
    """
    mask_rle: run-length as string formated (start length)ы
    shape: (width,height) of array to return
    Returns numpy array, 1 - mask, 0 - background
    """
    if mask_rle != mask_rle:
        return np.zeros_like(shape)

    s = mask_rle.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
    starts -= 1
    ends = starts + lengths
    img = np.zeros(shape[0]*shape[1], dtype=np.uint8)
    for lo, hi in zip(starts, ends):
        img[lo:hi] = 1
    return img.reshape(shape).T

class TrainDataset():
    # немного переписанный загрузчик данных опять же с конкурса
    def __init__(self, image_dir, train):
        self.image_dir = image_dir
        self.train = train
        self.image_uses = []

    def __getitem__(self, index):
        
        
        img_name = self.train.iloc[index]['ImageId']
        if img_name in self.image_uses:
            return 'hups'
        else:
            self.image_uses.append(img_name)
            path = os.path.join(self.image_dir, img_name)
            img = cv2.imread(path)
            img = cv2.resize(img,(384,384))

            ce_mask = [
                (i + 1) * rle2mask(rle, shape=(img.shape[0], img.shape[1])) \
                    for i, rle in enumerate(self.train[self.train['ImageId']==img_name]['EncodedPixels'])
            ]
            ce_mask = np.sum(ce_mask, axis=0, dtype=np.float32)


            result = {
                'image':img, 
                'mask': ce_mask,
                'pick_name':img_name
            }
            return result
    
    def __len__(self, ):
        return len(self.train)
    
train = pd.read_csv(MASK_PATH)
train_dataset = TrainDataset(
    image_dir=IMAGE_PATH, 
    train=train)
```

## Метрика dice

В самой задачке на бустере используется метрика dice по границе и всей области. Мы упростим задачу, просто берем dice по области.

Собственно нам надо реализовать следующую формулу

𝑑𝑖𝑐𝑒=∑𝑘=07(2∗|𝑝𝑟𝑒𝑑𝑖𝑐𝑡𝑖𝑜𝑛∩𝑡𝑟𝑢𝑐ℎ||𝑝𝑟𝑒𝑑𝑖𝑐𝑡𝑖𝑜𝑛|+|𝑡𝑟𝑢𝑐ℎ|)
 
Ну и чтобы в случае чего не делить на ноль, надо добавить небольшой эпсилон к верхней и нижней части.

Метрика dice. Так себе вариант - точный, но не дифференцируемый

```python
def dice_metric(y_true,y_pred):
    # по доке мы принимаем на вход 2 тензора
    # отдавать же должны скаляр
    classes=[0,1,2,3,4,5,6,7]
    eps = 0.1
    score=0
    for i in classes:
      cl=tf.constant(i,tf.float32) 
      y_true_cl = tf.math.equal(y_true,cl)
      y_pred_cl = tf.math.equal(y_pred,cl)
      y_true_cl = tf.dtypes.cast(y_true_cl,tf.float32)
      y_pred_cl = tf.dtypes.cast(y_pred_cl,tf.float32)

      intersection =tf.reduce_sum( tf.math.multiply(y_true_cl, y_pred_cl))
      score = score+ (2. * intersection + eps) / (tf.reduce_sum(y_true_cl)+ tf.reduce_sum(y_pred_cl) + eps)

    return score
 
```

Метрика dice. Рабочий вариант

```python
# Имеет ясный смысл для бинарных 0 1 тензоров
def dice_metric(y_true,y_pred):
    # по доке мы принимаем на вход 2 тензора
    # отдавать же должны скаляр
  eps = tf.constant(0.1,tf.float32)
  intersection =tf.reduce_sum( tf.math.multiply(y_true, y_pred))
  score =(tf.constant(2,tf.float32) * intersection + eps) / (tf.reduce_sum(y_true) + tf.reduce_sum(y_pred) + eps)

  return score
```

функция Тензор картинок делим на 255 X_rescale_255

```python
def X_rescale_255(X):
  return tf.math.divide(X, tf.constant(255,tf.float32))
```

функция Тензор масок в one-hot вид get_y_train_8

```python
def get_y_train_8(flowy):
  # flowy - тензор размера (sample,h,w), маска 8 классов
  # добавляем измерение размера 8
  y_train_cat = []
  
  for i in range(0,flowy.shape[0]):
    img_y_cat=np.zeros((flowy.shape[1],flowy.shape[2],8))
    for col in range(flowy.shape[2]):
      y_col=flowy[i,:,col]
      img_y_cat[:,col,:]=tf.keras.utils.to_categorical(y_col)
    y_train_cat.append(img_y_cat)
  y_train_cat=tf.convert_to_tensor(np.asarray(y_train_cat), dtype=tf.float32)
  return y_train_cat
```


## Предсказание - маска из выхода модели model_predict_mask

### Генератор одного потока с аугментацией и преобразованиями для zip картинок и масок

```python
class My_DataGenerator_1(tf.keras.preprocessing.image.ImageDataGenerator):


    def flow(self,
             x,
             y=None,
             batch_size=32,
             shuffle=False,
             sample_weight=None,
             seed=17,
             save_to_dir=None,
             save_prefix='',
             save_format='png',
             subset=None,
             expand_8=False,
             rescale_255=False):     

        for flowx in super().flow(x, y=None, batch_size=batch_size,
                                         shuffle=shuffle):
          if expand_8:
            flowx=flowx[:,:,:,0]
            
            flowx = get_y_train_8(flowx)
            
          if rescale_255:
            flowx=X_rescale_255(flowx)
          yield flowx
          #return flowx, y_train_cat
```

##  Простая Fully convolution NN  Имя: model_fully

```python
TARGET_WIDTH , TARGET_HEIGHT, IMG_CHANNELS = X_train.shape[1:]

# Вход в сетку 
inputs = Input((TARGET_WIDTH , TARGET_HEIGHT, IMG_CHANNELS))

# 128
down1 = Conv2D(8, (3, 3), activation="relu", padding='same')(inputs)
down1 = BatchNormalization()(down1)
down1 = Conv2D(8, (3, 3), activation="relu", padding='same')(down1)
down1 = BatchNormalization()(down1)
down1_pool = MaxPooling2D((2, 2), strides=(2, 2))(down1)

# 64
down2 = Conv2D(16, (3, 3), activation="relu", padding='same')(down1_pool)
down2 = BatchNormalization()(down2)
down2 = Conv2D(16, (3, 3), activation="relu", padding='same')(down2)
down2 = BatchNormalization()(down2)
down2_pool = MaxPooling2D((2, 2), strides=(2, 2))(down2)

# 32
center = Conv2D(32, (3, 3), activation="relu", padding='same')(down2_pool)
down3 = BatchNormalization()(center)
center = Conv2D(32, (3, 3), activation="relu", padding='same')(down3)
center = BatchNormalization()(center)

up2 = UpSampling2D((2, 2))(center)
up2 = Conv2D(16, (3, 3), activation="relu", padding='same')(up2)
up2 = BatchNormalization()(up2)
up2 = Conv2D(16, (3, 3), activation="relu", padding='same')(up2)
up2 = BatchNormalization()(up2)
up2 = Conv2D(16, (3, 3), activation="relu", padding='same')(up2)
up2 = BatchNormalization()(up2)

# 64
up1 = UpSampling2D((2, 2))(up2)
up1 = Conv2D(8, (3, 3), activation="relu", padding='same')(up1)
up1 = BatchNormalization()(up1)
up1 = Conv2D(8, (3, 3), activation="relu", padding='same')(up1)
up1 = BatchNormalization()(up1)
up1 = Conv2D(8, (3, 3), activation="relu", padding='same')(up1)
up1 = BatchNormalization()(up1)

# 128
outputs = Conv2D(8, (1, 1), activation='sigmoid')(up1)
model_fully = Model(inputs=inputs, outputs=outputs)
```

## Собираем u-net. Смотрим как оно :)

Unet, но попроще

```python
def conv2d_block(input_tensor, n_filters, kernel_size=3, batchnorm=True):

    # first layer
    x = Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size), kernel_initializer="he_normal",
               padding="same")(input_tensor)
    if batchnorm:
        x = BatchNormalization()(x)
    x = Activation("relu")(x)
    # second layer
    x = Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size), kernel_initializer="he_normal",
               padding="same")(x)
    if batchnorm:
        x = BatchNormalization()(x)
    x = Activation("relu")(x)
    return x

def get_unet(input_img, n_filters=8, dropout=0.2, batchnorm=True):

  # contracting path
  c1 = conv2d_block(input_img, n_filters=n_filters*1, kernel_size=3, batchnorm=batchnorm)
  p1 = MaxPooling2D((2, 2)) (c1)
  #p1 = Dropout(dropout*0.3)(p1)

  c2 = conv2d_block(p1, n_filters=n_filters*2, kernel_size=3, batchnorm=batchnorm)
  p2 = MaxPooling2D((2, 2)) (c2)
  #p2 = Dropout(dropout)(p2)

  c3 = conv2d_block(p2, n_filters=n_filters*4, kernel_size=3, batchnorm=batchnorm)
  
  u5 = Conv2DTranspose(n_filters*2, (3, 3), strides=(2, 2), padding='same') (c3)
  u5 = concatenate([u5, c2])
  #u5 = Dropout(dropout)(u5)
  u5 = conv2d_block(u5, n_filters=n_filters*2, kernel_size=3, batchnorm=batchnorm)

  u6 = Conv2DTranspose(n_filters*1, (3, 3), strides=(2, 2), padding='same') (u5)
  u6 = concatenate([u6, c1], axis=3)
  #u6 = Dropout(dropout)(u6)
  u6 = conv2d_block(u6, n_filters=n_filters*1, kernel_size=3, batchnorm=batchnorm)

  u6=  Conv2D(filters=n_filters, kernel_size=1, kernel_initializer="he_normal",
               padding='same')(u6)
  output = Activation('sigmoid')(u6)
  return output
```

```python
%matplotlib inline
def plot_loss(his, title):
    plt.style.use('ggplot')
    plt.figure()
    epoch=len(his.history['val_dice_metric'])
    plt.plot(np.arange(0, epoch), his.history['loss'], label='loss')
    plt.plot(np.arange(0, epoch), his.history['val_loss'], label='val_loss')
    
    plt.title(title)
    plt.xlabel('Epoch #')
    plt.ylabel('Loss')
    plt.legend(loc='upper right')
    plt.show()

def plot_dice(his,  title):
    plt.style.use('ggplot')
    plt.figure()
    epoch=len(his.history['val_dice_metric'])
    plt.plot(np.arange(0, epoch), his.history['dice_metric'], label='dice_metric')
    plt.plot(np.arange(0, epoch), his.history['val_dice_metric'], label='val_dice_metric')
   
    plt.title(title)
    plt.xlabel('Epoch #')
    plt.ylabel('Dice metric')
    plt.legend(loc='upper right')
    plt.show()
```

```python
for his,tit in zip(hists,titles):
  plot_dice(his,  tit)
  plot_loss(his, tit)
```
